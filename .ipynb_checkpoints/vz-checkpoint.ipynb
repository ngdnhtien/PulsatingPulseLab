```python
from qiskit import pulse
from qiskit import IBMQ
import warnings
warnings.filterwarnings('ignore')

IBMQ.load_account() # load if no backend satisfies
provider = IBMQ.get_provider(hub='ibm-q', group='open', project='main')
backend = provider.get_backend('ibm_oslo')
qubit = 0
backend_config = backend.configuration()
dt = backend_config.dt  
backend_defaults = backend.defaults()
```

```python
f01 = 4.92517*1.0e9
f12 = 4581552617.648945

d12_duration = 544
d12_amplitude = 0.10879003883868105
d12_sigma = 40

d01_duration = 544
d01_sigma = 67
d01_amplitude = 0.07999888439750123
```

```python
phi = Parameter('phi')
with pulse.build(backend=backend, default_alignment='sequential', name=r'$X_{\pi}^{(01)}$') as xp01_inst:
    drive_duration = d01_duration
    drive_sigma = d01_sigma
    pi_amp = d01_amplitude
    drive_chan = pulse.drive_channel(qubit)
    pulse.set_frequency(f01, drive_chan)
    with pulse.phase_offset(phi, drive_chan):
        pulse.play(pulse.Gaussian(duration=drive_duration,
                                  amp=pi_amp,
                                  sigma=drive_sigma,
                                  name=r'$X_{\pi}^{(01)}$'), drive_chan)

xp01_inst.draw()
```

```python
with pulse.build(backend=backend, default_alignment='sequential', name=r'$X_{\pi}^{(12)}$') as xp12_inst:
    drive_chan = pulse.drive_channel(qubit)
    pulse.set_frequency(f12, drive_chan)
    pulse.play(pulse.Gaussian(duration=d12_duration,
                              amp=d12_amplitude,
                              sigma=d12_sigma,
                              name=r'$X_{\pi}^{(12)}$'), drive_chan)
xp12_inst.draw()
```

```python
from qiskit.circuit import QuantumCircuit, Gate

xp01 = Gate(r'Xp01', 1, [])
xp12 = Gate(r'Xp12', 1, [])

qc = QuantumCircuit(1,1)
qc.append(xp01, [0])
qc.append(xp01, [0])
qc.append(xp01, [0])
qc.append(xp12, [0])
qc.append(xp12, [0])
qc.measure(0,0)
qc.add_calibration(xp01, (0,), xp01_inst)
qc.add_calibration(xp12, (0,), xp12_inst)
```

```python
qc.draw()
```

```python
from qiskit import schedule

qc_sched = schedule(qc, backend)
qc_sched.draw(backend=backend)
```

### Creating $Y_\theta^{(01)}$ pulses on computational subspaces

Given that an arbitrary pulse on superconducting qutrit's (mn) subspace is specified by 

\begin{align}
U = \exp\left[-i\dfrac{\theta}{2}\left(\cos(\phi)\lambda_x^{(mn)}+\sin(\phi)\lambda_y^{(mn)}\right)\right]
\end{align}

where the generators of rotation are the Gell-Mann matrices specified [here](https://github.com/ngdnhtien/PulsatingPulseLab/blob/main/exp_gell.ipynb). Let us now put into test one scenario where we have to create a sequence of gates, for example

\begin{align}
Y^{(01)}_{\theta}X^{(01)}_{\theta}Z^{(01)}_{\phi}X^{(01)}_\theta
\end{align}

First principle calculation from McKay (2017) shows that if we were to *add a phase offset $-\phi$* to the second pulse, i.e. the $X_{\theta}^{(01)}$, we we'll be able to create 

\begin{align}
Z^{(01)}_{-\phi}X_{\theta}^{(01)}Z^{(01)}_{\phi}X^{(01)}_\theta
\end{align}

Let us see that sequence in action!

```python
from qiskit.circuit import QuantumCircuit, Gate, Parameter
import numpy as np

polish = [0, np.pi/2]
xp01_phi = Gate('Xp01_phi', 1, [phi]) # define a gate for our xp01 pulse with phase offset = phase_offset_val
xp01 = Gate('Xp01', 1, [])

qc = QuantumCircuit(1,1) # let us create a quantum circuit with 1 logical qubit and 1 classical bit.
qc.append(xp01, [0])
qc.append(xp01_phi, [0])
qc.add_calibration(xp01, (0,), xp01_inst, [])
qc.add_calibration(xp01_phi, (0,), xp01_inst, [phi])
```

```python
from qiskit import schedule

qc_sched = schedule(qc, backend)
qc_sched.draw(backend=backend)
```
